```python
import pandas as pd
import numpy as np
```

```python
#データの読み込み
teams=pd.read_csv("teams.csv")
boxscore=pd.read_csv("games_boxscore_202021.csv")
```

```python
#表示される行・列の数を増やす
pd.set_option("display.max_columns",100)
pd.set_option("display.max_rows",1000)
```

```python
#2020-21シーズンのB1のデータがほしいのでteamsから絞り込む
teams=teams[teams["Season"]=="2020-21"]
teams=teams[teams["League"]=="B1"]
```

```python
#teamsをmergeしてboxscoreからB1のデータを絞り込む
boxscore=pd.merge(boxscore,teams,on="TeamId")
```

```python
#選手の平均スタッツを計算する
boxscore=boxscore.groupby("Player").mean()
boxscore=boxscore.reset_index()
```

```python
#「バスケのスタッツを眺めるアプリ」で使ったスタッツを抽出（OE,eFG,TS%,PPPは除く）
boxscore=boxscore.loc[:,["Player","MIN","PTS","FGM","FGA","F3GM","F3GA","FTM","FTA","OR","DR","TR","AS","TO","ST","BS","EFF"]]
```

```python
#相関行列を出力
boxscore.drop("Player",axis=1).corr()
```

```python
#偏相関係数を算出する準備として、PTS,TR,AS,TOの各組み合わせの相関係数を保存しておく
r_pts_tr=boxscore["PTS"].corr(boxscore["TR"])
r_pts_as=boxscore["PTS"].corr(boxscore["AS"])
r_pts_to=boxscore["PTS"].corr(boxscore["TO"])

r_tr_as=boxscore["TR"].corr(boxscore["AS"])
r_tr_to=boxscore["TR"].corr(boxscore["TO"])

r_as_to=boxscore["AS"].corr(boxscore["TO"])
```

```python
#偏相関係数を算出する準備として、PTS,TR,AS,TOとMINの相関係数を保存しておく
r_min_pts=boxscore["MIN"].corr(boxscore["PTS"])
r_min_tr=boxscore["MIN"].corr(boxscore["TR"])
r_min_as=boxscore["MIN"].corr(boxscore["AS"])
r_min_to=boxscore["MIN"].corr(boxscore["TO"])
```

```python
#偏相関係数の算出
#PTSとTR
(r_pts_tr-(r_min_pts*r_min_tr))/(np.sqrt(1-r_min_pts**2)*np.sqrt(1-r_min_tr**2))
```

```python
#偏相関係数の算出
#PTSとAS
(r_pts_as-(r_min_pts*r_min_as))/(np.sqrt(1-r_min_pts**2)*np.sqrt(1-r_min_as**2))
```

```python
#偏相関係数の算出
#PTSとTO
(r_pts_to-(r_min_pts*r_min_to))/(np.sqrt(1-r_min_pts**2)*np.sqrt(1-r_min_to**2))
```

```python
#偏相関係数の算出
#TRとAS
(r_tr_as-(r_min_tr*r_min_as))/(np.sqrt(1-r_min_tr**2)*np.sqrt(1-r_min_as**2))
```

```python
#偏相関係数の算出
#TRとTO
(r_tr_to-(r_min_tr*r_min_to))/(np.sqrt(1-r_min_tr**2)*np.sqrt(1-r_min_to**2))
```

```python
#偏相関係数の算出
#ASとTO
(r_as_to-(r_min_as*r_min_to))/(np.sqrt(1-r_min_as**2)*np.sqrt(1-r_min_to**2))
```
